```python
#List#

numbers=[10,20,30,40,50]
print(len(numbers))

#adding elements using append#
numbers=[100,200,300,400]
numbers.append(500)
print(numbers)

#removing an elemet using pop#
numbers=[100,200,300,400,500]
numbers.remove(300)
print(numbers)


#sorting numbers in list#
numbers=[9,8,6,7,1,3,2,5,]

numbers.sort()
print(numbers)

#finding the maximum and mininum elements#
numbers=[1000,900,800,700,600,500]
print(min(numbers))
print(max(numbers))



```

```python
#tuple#

t1=(1,23,35,45,55,65,77)
print(len(t1))

#access elements using index#
t=(10,20,30,40,50)
print(t[4])
print(t[3])
print(t[2])

#count the occurence of elements#
t=(101,102,103,104,105,106,107,102,108,102,109,102)
print(t.count(102))

#index of element (changed to a value present in the tuple, e.g., 4000)#
t=(1000,2000,3000,4000,5000)
print(t.index(4000))

#find minimum and maximum value (corrected function calls)#
t=(2000,3000,4000,5000,6000)
print(min(t))
print(max(t))
```

```python
#dictionary#
student={"name":"reddy","age":20,"city":"hyd","course":"python","grade":"A"}
print(student)

#accessing value using a key#
student={"name":"reddy","age":20,"course":"python"}
print(student["course"])

#using get()#
student={"name":"reddy","age":20,"course":"python"}
print(student.get("age"))


#add or update a key-value pair#
student={"name":"reddy","age":20}
student["grade"]="A"
student["age"]=21
print(student)
```

### The `for` loop

#### Iterating over a list

```python
# Iterate over a list of fruits
fruits = ["apple", "banana", "cherry",]
for x in fruits:
  print(x)
```

#### Looping through a string

```python
# Loop through the letters in a string
for i in "banana":
  print(i)
```

```python
#break statement#
# Exit the loop when x is "banana"
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
#while loop#
n=int(input("enter a number"))
i=1
while i<=20:
    print("n,", "*" * i, i, ")")
    i=i+1
```

#### The `break` statement

With the `break` statement we can stop the loop even if the while condition is true:

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

### Conditional Statements (`if`, `elif`, `else`)

Conditional statements allow you to execute different blocks of code based on whether a condition is true or false.

```python
#if statement#
a = 33
b = 200

if b > a:
  print("b is greater than a")
```

```python
#elif statement#
a = 33
b = 33

if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

```python
#else statement#
a = 200
b = 33

if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```
